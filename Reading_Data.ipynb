```python
import pandas as pd
import gzip

def parse(path):
    g = gzip.open(path, 'rb')
    for l in g:
        yield eval(l)

def getDF(path):
    i = 0
    df = {}
    positive_count = 0
    negative_count = 0
    neutral_count = 0
    for d in parse(path):
        if d['overall'] >= 4.0:
            if positive_count < 40000:
                positive_count += 1
                df[i] = d
                if positive_count % 10000 == 0:
                    print('positive: ', positive_count)
        elif d['overall'] == 3.0:
            if neutral_count < 40000:
                neutral_count += 1
                df[i] = d
                if neutral_count % 10000 == 0:
                    print('neutral: ', neutral_count)
        else:
            if negative_count < 40000:
                negative_count += 1
                df[i] = d
                if negative_count % 10000 == 0:
                    print('negative: ', negative_count)
        i += 1
        if positive_count + neutral_count + negative_count >= 40000 * 3:
            break
    return pd.DataFrame.from_dict(df, orient='index')
```

```python
df = getDF('reviews_Electronics.json.gz')
```
